add_axes([a,b,c,d])

* a is the x origin of the plot
* b is the y origin of the plot
* c is the width
* d is the height

All numbers are in the range [0,1] (0 to 1, both included)

```python
figure = plt.figure(figsize=(10,8))
s1 = figure.add_axes([0.5,0.5,0.25,0.25])
s1.plot(x,y1,'r')
s2 = figure.add_axes([0.1,0.1,0.2,0.1])
s2.plot(x,y2,'b')
plt.show()
```

Determine the probability distribution function for a  tt  statistic with 15 degrees of freedom and save it in a variable called t_dist. Next, calculate the cumulative distribution function for t_dist with t_values and save it in a variable called t_cdf.

```python
import scipy
```

```python

# Calculate all below with this values
t_values = np.linspace(-4, 4, 1000)

### YOUR SOLUTION HERE
t_dist = scipy.stats.t(15).pdf(t_values)
t_cdf = np.cumsum(t_dist)
```
